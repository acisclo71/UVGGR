```python
syms=sympy.symbols('r theta phi')
```

```python
#### Definir la metrica para coordenadas esféricas
metric=[[0 for i in range(3)] for i in range(3)]
metric[0][0]=1
metric[1][1]=syms[0]**2
metric[2][2]=(syms[0]**2)*(sympy.sin(syms[1])**2)
m_obj=MetricTensor(metric,syms)
```

```python
m_obj.tensor()
```

```python
chris=ChristoffelSymbols.from_metric(m_obj)
```

```python
chris.tensor()
```

```python
chris.tensor()[2,2,1]
```

```python
rm1 = RiemannCurvatureTensor.from_christoffels(chris)
```

```python
rm1.tensor()
```
